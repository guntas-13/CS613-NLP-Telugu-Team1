### Scrapping the Website :- https://ambuda.org/

```python
from selenium import webdriver
import time
import pprint
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get("https://ambuda.org/texts")
```

```python
classes = driver.find_elements(by = By.CSS_SELECTOR , value=".py-1.block.text-xl.hover\:underline")
```

```python
links = []
```

```python
for cls in classes:
    cls.click()
    time.sleep(0.1)
    articles = driver.find_elements(by=By.CSS_SELECTOR, value=".block.hover\:underline")
    
    for article in articles:
        links.append(article.get_attribute("href"))
        time.sleep(0.1)

    driver.back()
    time.sleep(0.1)
```

```python
driver.quit()
```

```python
pprint.pprint(links)
```

```python
import os

os.makedirs("ambuda", exist_ok=True)
file = open("ambuda/links.txt", "w")
for link in links:
    file.write(link + "\n")
```

### Get text data from the source

```python
links = open("ambuda/links.txt", "r").read().split("\n")
```

```python
driver = webdriver.Chrome()
```

```python
for link in links:
    print(link)
    driver.get(link)
    time.sleep(0.1)
    temp = link.split("/")
    title = temp[4] + "_" + temp[5]
    title = title.replace(".", "_")
    file = open("ambuda/" + title + ".txt", "w")

    texts = driver.find_elements(by=By.CSS_SELECTOR, value="s-l")

    for text in texts:
        if text.text != "":
            file.write(text.text + "\n")

    file.close()
```

```python
driver.quit()
```
